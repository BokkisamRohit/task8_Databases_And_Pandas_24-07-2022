```python
import pandas as pd
import mysql.connector as conn
from sqlalchemy import create_engine
class MysqlConnection:
    __host=''
    __password=''
    __userName=''
    __databaseName=''
    
    def setVariables(self,host,password,userName,databaseName):
        logging.info("The given host name is %s", host)
        self.__host=host
        #We should never log a password info
        self.__password=password
        logging.info("The given user's name is %s", userName)
        self.__userName=userName
        logging.info("The given database's name is %s", databaseName)
        self.__databaseName=databaseName    
    def createMySQLConnection(self):
        return conn.connect(host=self.__host,user=self.__userName,passwd=self.__password,database=self.__databaseName)
    
    def createMySQLEngine(self,portNum):
        return create_engine('mysql+mysqlconnector://{0}:{1}@{2}:{3}/{4}'.format(self.__userName,self.__password,self.__host,portNum,self.__databaseName))
    
    def createMySQLCursor(self,mydbcon):
        return mydbcon.cursor()
    
    def createMySQLCursorExecuteObject(self,cur,q):
        return cur.execute(q)
```

```python
m=MysqlConnection()
host=input("Enter the host: ")
password=input("Enter the password: ")
userName=input("Enter the user's name: ")
databaseName=input("Enter the database's name: ")
portNumber=input("Enter the port number: ")
m.setVariables(host,password,userName,databaseName)
try:
    mydb=m.createMySQLConnection()
    logging.info("The given connection status %s",mydb)
    cursor=m.createMySQLCursor(mydb)
    logging.info("The given cursor is %s",cursor)
    engine=m.createMySQLEngine(portNumber)
    logging.info("The given MySQL Engine is %s",engine)
except Exception as e:
    logging.exception("The exception that we have got:" + "\n" + str(e))
```
